```python
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET
import requests
import pandas as pd
```

* 한국 소설 판매량 내림차순으로 100개

```python
url = 'https://book.interpark.com/api/search.api?key=7038BCF1AD98D760E4576993E36C60F5714F4DF2F2A865235C5905F51948DD7E&categoryId=101&maxResults=100&sort=salesPoint&query=%EC%86%8C%EC%84%A4'
```

```python
req = requests.get(url)
xml = req.text
soup = BeautifulSoup(xml, 'xml')
```

### api에서 수집할 목록 = name

```python
name = ('itemId', 'title', 'pubDate', 'coverSmallUrl', 'coverLargeUrl', 'publisher', 'customerReviewRank', 'author', 'isbn', 'link')
```

#### 태그 찾아서 리스트에 추가해주는 함수 : 수집한 100개 목록(101개는 0번 인덱스가 다른것 확인 완료)

```python
def find_tag_n_make_ls(find_tag, ls_name):
    find_ls = soup.find_all(find_tag)
    if len(find_ls) == 101:
        for i in find_ls[1:]:
            ls_name.append(i.text)
    elif len(find_ls) == 100:
        for i in find_ls:
            ls_name.append(i.text)
    else:
        print('100개 아님 확인 요망!')
        return 
        
    return ls_name
```

```python
all_ls = list()
for i in range(len(name)):
    erase_ls = list()
    itemId_ls = find_tag_n_make_ls(name[i], erase_ls)
    all_ls.append(itemId_ls)
    print(len(itemId_ls))
# all_ls
```

---

```python
df_dict = dict()
for i in range(len(all_ls)):
    df_dict[name[i]] = all_ls[i]
df = pd.DataFrame(df_dict)
```
