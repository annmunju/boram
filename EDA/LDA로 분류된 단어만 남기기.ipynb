```python
import pandas as pd
```

```python
df = pd.read_csv('소설_책내용_서평기반_LDA_완료.csv', index_col=0)
```

```python
df[df['Perc_Contribution']>0.8]
```

```python
story = df['story']
```

```python
stories = []
for i in range(len(story)):
    stories.append(story[i].split())
stories
```

```python
all_story = []
for i in range(len(stories)):
    all_story += stories[i]
```

```python
topic_keywords = set()
for tk in df['Topic_Keywords'].unique():
    topic_keywords.update(list(tk.split(', ')))
print(topic_keywords)
```

```python
len(topic_keywords)
```

```python
need_story = []
for a_s in all_story:
    if a_s in topic_keywords:
        need_story.append(a_s)
need_story
```

```python
import collections
counts = collections.Counter(need_story)
```
