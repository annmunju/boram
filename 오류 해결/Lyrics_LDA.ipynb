```python
preprocessedLy = song_chart['preprocessed']

pre_ly = []
for i in range(len(preprocessedLy)):
    ly_list = preprocessedLy[i][2:-2].split("', '")
    pre_ly.append(ly_list)


pre_ly
```

```python
import gensim
from gensim import corpora
from gensim import models
from gensim.models import LdaModel
from gensim.models import TfidfModel
from gensim.models import CoherenceModel
from gensim.models.wrappers import LdaMallet
```

```python
id2word=corpora.Dictionary(pre_ly)
id2word.filter_extremes(no_below = 20) #20회 이하로 등장한 단어는 삭제
texts = pre_ly
corpus=[id2word.doc2bow(text) for text in texts]
```

```python
mallet_path = './mallet-2.0.8/bin/mallet'
ldamallet = LdaMallet(mallet_path, corpus=corpus, num_topics=3, id2word=id2word)
```
